#No persona

IR_baseline

```python
from parlai.scripts.train_model import TrainModel
TrainModel.main(
    # we MUST provide a filename
    model_file='no_persona_trained_models/ir_baseline' ,
    # train on persona dialogues
    task='personachat:none',
    # limit training time to 5 minutes, and a batchsize of 16
    max_train_time=5 * 60,
    batchsize=16,
    
    # we specify the model type as ir_baseline model
    model='ir_baseline',
)
```

```python
from parlai.scripts.display_model import DisplayModel
DisplayModel.main(
    task='personachat',
    model_file='no_persona_trained_models/ir_baseline',
    num_examples=100,
)
```
